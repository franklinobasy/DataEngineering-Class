```python
dd = df_dim[['GeographyKey', 'City', 'StateProvinceCode', 'StateProvinceName', 'IpAddressLocator']]
dd.iloc[2].values
```

```python
cols = dd.columns
```

```python
dd.head()
```

```python
data_randon = [[6, 9], ["paris", "monaco"], ["EWS", "SWE"], ["West France", "North France"], ["12.17.10.1", "17.17.8.1"]]
french_df = pd.DataFrame(data_randon).T
french_df
```

```python
french_df.columns = cols
french_df
```

```python
####Append

appended = dd.append(french_df)
df_r = appended.reset_index()
df_p = df_r.drop("index", axis=1)
df_p
```

```python
###concat 

new_con = pd.DataFrame([55, 66, 11, 12, 8, 19, 13], columns=["Weather"])
new_con
```

```python
df_concat = pd.concat([df_p, new_con], axis=1)
df_concat.head()
```

```python
ip_list = []
for i in df_concat["IpAddressLocator"].values:
    if i.split(".")[0] == '198':
        ip_list.append("British Ip")
    else:
        ip_list.append("French_Ip")
df_concat["Iplocation"] = ip_list

df_concat
```
